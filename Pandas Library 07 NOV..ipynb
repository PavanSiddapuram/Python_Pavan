# Labels
If nothing else is specified,the values are labeled with their index number.first value has index(0),sencond value has index 1 etc..

This label can be used to access a specified value.

```python
import pandas as pd
c = [20,22,24]
myvar = pd.Series(c,index=["x","y","z"])
print(myvar)
```
